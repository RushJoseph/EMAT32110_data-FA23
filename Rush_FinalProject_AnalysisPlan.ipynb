## Assumptions
- Standard precision team comp: 1 commando, 2 gunslingers, 2 sharpshooters, as well as 1 medic
- Meta weapons
 - 9mm, Dual 1858, SPX(+2), dual deagles, and single AF for gunslinger
 - 9mm, Winchester, EBR(+1), SPX(+2) for sharpshooter
 - 9mm, AR15, Bullpup, SCAR, FN FAL for commando
- Unmodded weapons
- Weapon damage modeled at a consistent 5 REU. This isn’t quite accurate to real play, particularly for players with lower accuracy or playing parts of the map where they don’t have a constant flow of enemies to shoot, but it’s close enough to reality for this model. A more thorough model would look at shots fired per wave and real wave durations and perhaps model REU based on how many 2 second intervals go by without an headshots expected to be hit 
- Sharpshooters use 4 freeze grenades each in every wave. This is very realistic to actual play. If anything, they tend to use fewer. 
- All players only fill SPX to 80 rounds
- All zeds are killed by players
- Medics: see Tradeoff #1

#### Sharpshooter Damage Skills

- RLLRR skills

## Tradeoffs:
1.	Headshot damage only. It is often reasonable to kill the weak enemies in-game with body shots. Hit zone data does exist for zed kills, but it does not account for prior shots or, in the case of larges, for decapitations before the body was killed. A fleshpound cannot be assumed to have been killed with body shots when it was actually decapitated first and then shot in the chest for the quadruple decapped body multiplier. Additionally, precision play emphasizes headshots, as they are most effective, and for the sake of not expanding the model into intractably unknowable territory, only headshots will be applied to zeds.
2.	Medics excerpted, and they are an extremely valuable part of team eco. They will be assumed for any amount of dosh required for both sharps to load up on w2 EBR and freezes. To include medics, I would need real world wave by wave eco data for them, and I don’t have that. The medic is then assumed to be able to contribute that much dosh to the team’s pool each wave.
3.	Assumes that overkill (wasted) damage and penetration damage roughly cancel each other out. This is not necessarily true in practice, but there is no good way to track them available either in-game or in the accumulated data.
4.	Weighting for zed damage resistances would be best derived from actual wave by wave kills per perk per cycle, but that’s beyond the scope of the easily accessible data.
5.	Only roughly accounts for target prioritization
6.	Currently does not account for shared money from kills. I might distribute 40% of large kill dosh across the trinity (mando, sharp 1, slinger 1) to account for some of this, roughly modeling Nausea
7.	Grenade damage for mando & gunslinger not included, since it is incredibly variable
8.	Dead-Eye ignored for sharpshooter. Its impact is unreliable, since it depends on individual playstyles, maps, and situations, so its existence is being disregarded.
9.	Cost per bullet isn’t quite exact to in-game modeling, since I used cost per magazine and there’s a little more stuff happening under the hood than that, but it’s close enough for this.
10.	The assumption that all zeds are killed by players is inaccurate to Friendly Fire, particularly in the case of Fleshpounds dealing damage to other zeds while raging or raged. However, zeds also have a chance to rout around the fleshpounds or move out of their way, so the modeling for this is intractable without massive amounts of game data of per wave zed kills by players on any particular cycle. The difference between the population and the players’ kills would give the number of zeds killed by friendly fire.

## Modeling
Presentation of damage formula
### Data gathering
1.	Gather acc & hsacc from summer tournament & EU NCZ group. All sharps & slingers in one group, all mandos in another. Trim out Tamariland significant AF users due to data distortion. (Significant defined as them buying dual AFs)
2.	Gather final zed kill totals per player from tournament data. Separate into two pools: sharp/slinger & mando
3.	Final damage per player. Per wave, if available.
4.	Zed population data for bl_v2, nt_v2, osffi_v1, apm_v2, ts_mig_v2, doom_v2_plus_rmk, ts_mig_v3, nvm_v1
5.	Weapon stats 
 - Cost per weapon
 - Cost per bullet
 - Base damage per weapon
 - Cost per upgrade
 - Damage increase per weapon
 - Upgraded base damage per weapon per upgrade

### Data Analysis
1.	__**SCATTER PLOT**__ of acc vs. hsacc
2.	Group acc/hsacc data with analysis to make a few representative player skill levels.
3.	Find _Weighted Damage Resistance Values_ i.e. FPs are more resistant to handgun than rifle. Use percentage of a given zed in the wave averaged with percentage of overall game kills from tourney data for the weight of that zed’s resistance value. If time, use tourney data for kills per wave.
4.	Find the adjusted zed head health pool for a given wave. [Head health]/[headshot multiplier] for every zed. This results in a big number.
5.	Find also: adjusted zed head health pool for NVM_v1 w10. This acts as the cutoff for the eco model.
6.	Find damage per bullet for each weapon before zed resistances are accounted for with standard skills at 5 racks where REU is relevant
7.	Multiply by current ammunition size to get _Max Damage Per Weapon_
8.	Find the Max Damage per Player for a given wave. This is [damage for every available bullet in a weapon]*[weighted resistance value] + the same for all their other weapons.
9.	Add together all the _Max Damage per Player_ values to get _Max Damage for Team_.
10.	Use the Max Damage Per Weapon derive an Expected Damage Weight percentage.
11.	Multiply the above value by hsacc% of the sample skill level for each player to get Expected Damage Per Player.
12.	Add together the expected damage of each player to get _Expected Team Damage Pool_.
13.	Subtract _Adjusted Zed Head Health Pool_ and _Expected Team Damage_.
14.	__**BAR PLOT**__ of _Max Damage for Team_ vs. _Expected Team Damage_ vs. _Adjusted Zed Head Health Pool_.
15.	Use the damage per perk data, either per wave or final in the matches, to create a _Damage Pool Weight_ that represents each perk’s expected contribution to the total damage. This helps account for target priority, where the mando in particular is firing at many low health targets and largely ignoring high health targets.
16.	Divide the spent damage among the players using the _Damage Pool Weight__ and their weapons to get _Damage Per Player-_ and spread across their weapons to get _Expected Ammunition Expenditure_. (This will weight 9mm higher than actual play. That is an acceptable tradeoff)
17.	If a wave total wants too much damage from a player, such as a mando not having enough ammo to deal the expected damage, distribute the overflow among the other players equally.
18.	Use the _Damage per Player_ and their HSACC% to find their ammo expenditure
19.	Find the _Minimum Dosh Total_, which is the minimum amount of money the team can earn if every zed is killed with no teamwork.
20.	Divide the _Minimum Dosh Total_ among the players. Buy, fill, sell, & upgrade weapons from the model build orders.
21.	Use any leftover money from players to buy ammo, freeze grenades, & upgrades for other players (this will have to be by hand. Writing an optimization algorithm to maximize team damage output is beyond my current skills).
22.	Use any leftover money to buy grenades for non-sharpshooters
23.	Save any remaining money for future use
24.	Repeat with next wave
25.	_**LINE PLOT**_ of team money per wave.
26.	Conglomerate _**LINE PLOT**_ of team money per wave for different cycles.
27.	When _Expected Team Damage Pool_ is greater than _NVM_v1 w10 Adjusted Head Health Pool_, cut the model off. Players, if they play well, can deal with basically any number of zeds in waves after this point. It is the "Git Gud" Threshold.
